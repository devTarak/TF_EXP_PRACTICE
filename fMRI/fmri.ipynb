```python
import nibabel as nb

# Let's create a short helper function to plot 3D NIfTI images
def plot_slice(fname):

    # Load the image
    img = nb.load(fname)
    data = img.get_data()

    # Cut in the middle of the brain
    cut = int(data.shape[-1]/2) + 10

    # Plot the data
    imshow(np.rot90(data[..., cut]), cmap="gray")
    gca().set_axis_off()
```

```python
my_path="/home/bhaskard/myfolders/fmri/datasets/ds000007_R2.0.1/sub-01/anat"

from nipype.interfaces import fsl

# Skullstrip process
skullstrip = fsl.BET(
    in_file=my_path + "/sub-01_T1w.nii.gz",
    out_file=my_path + "/output/sub-01_T1w_brain.nii.gz",
    mask=True)
skullstrip.run()

# Smoothing process
smooth = fsl.IsotropicSmooth(
    in_file=my_path + "/sub-01_T1w.nii.gz",
    out_file=my_path + "/output/sub-01_T1w_smooth.nii.gz",
    fwhm=4)
smooth.run()

# Masking process
mask = fsl.ApplyMask(
    in_file=my_path + "/output/sub-01_T1w_smooth.nii.gz",
    out_file=my_path + "/output/sub-01_T1w_smooth_mask.nii.gz",
    mask_file=my_path + "/output/sub-01_T1w_brain_mask.nii.gz")
mask.run()

f = plt.figure(figsize=(12, 4))
for i, img in enumerate(["T1w", "T1w_smooth",
                         "T1w_brain_mask", "T1w_smooth_mask"]):
    f.add_subplot(1, 4, i + 1)
    if i == 0:
        plot_slice("/home/bhaskard/myfolders/fmri/datasets/ds000007_R2.0.1/sub-01/anat/output/sub-01_%s.nii.gz" % img)
    else:
        plot_slice("/home/bhaskard/myfolders/fmri/datasets/ds000007_R2.0.1/sub-01/anat/output/sub-01_%s.nii.gz" % img)
    plt.title(img)
```
